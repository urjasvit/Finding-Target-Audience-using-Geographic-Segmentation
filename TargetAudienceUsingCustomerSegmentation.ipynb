For starters we will first import pandas and numpy libraries for data wrangling tasks. After this we will load the relevant files to their dataframes and filter out the columns required to perform necessory tasks on the dataset.


```python
import numpy as np
import pandas as pd
```

```python
SexbyAge_df= pd.read_excel("D:/SexByAge(seq2).xlsx",header=1)
SexbyAge_df.columns.tolist()
```

```python
SexbyAge_df= pd.read_excel("D:/SexByAge(seq2).xlsx",header=1)
SexbyAge_df= SexbyAge_df.drop(SexbyAge_df.columns[6:99], axis=1)
SexbyAge_df= SexbyAge_df.drop(SexbyAge_df.columns[36:], axis=1)
SexbyAge_df.head()
```

```python
Race_df= pd.read_excel("D:/Race(seq3).xlsx",header=1)
Race_df= Race_df.drop(Race_df.columns[17:], axis=1)
Race_df= Race_df.drop(Race_df.columns[Race_df.columns.get_loc('DETAILED RACE% Total population%Total')],axis=1)
Race_df.shape
```

```python
MedianHousehold_df= pd.read_excel("D:/MedianHouseholdIncome(seq77).xlsx", header=1)
MedianHousehold_df= MedianHousehold_df.iloc[0:,[5,176]]
MedianHousehold_df.shape
```

```python
TypesofHealthInsuarance_df= pd.read_excel("D:/TypesOfHealthInsuranceCoverageByAge(171).xlsx", header=1)
TypesofHealthInsuarance_df= TypesofHealthInsuarance_df.drop(TypesofHealthInsuarance_df.columns[6:120],axis=1)
TypesofHealthInsuarance_df= TypesofHealthInsuarance_df.drop(TypesofHealthInsuarance_df.columns[72:],axis=1)
TypesofHealthInsuarance_df.head()
```

Creating merged_df dataframe to get all the required columns from each excel and merge them into one using left join.

```python
merged_df= pd.merge(left= SexbyAge_df, right=Race_df, how='left', left_on='LOGRECNO', right_on='LOGRECNO')
merged_df.head()
```

Filtering out unwanted columns using regular expression.

```python
merged_df.drop(merged_df.filter(regex='_y$|_x$').columns.tolist(),axis=1, inplace=True)
merged_df.head()
```

Merging the MedianHousehold_df to the merged_df dataframe using left join on LOGRECNO.

```python
merged_df= pd.merge(left= merged_df, right=MedianHousehold_df, how='left', left_on='LOGRECNO', right_on='LOGRECNO')
merged_df.head()
```

Filtering out non-required columns from the newly merged dataframe

```python
merged_df= pd.merge(left= merged_df, right=TypesofHealthInsuarance_df, how='left', left_on='LOGRECNO', right_on='LOGRECNO')
merged_df.drop(merged_df.filter(regex='_y$|_x$').columns.tolist(),axis=1, inplace=True)
merged_df.head()
```

Reading the Illinois state data from the 1_year_Mini_Geo.xlsx and saving it to dataframe

```python
Geography_df= pd.read_excel("D:/1_year_Mini_Geo.xlsx", sheet_name='il')
Geography_df.dtypes
```

Merging Geography_df with the merged_df using left join on LOGRECNO and Logical Record Number

```python
merged_df= pd.merge(left= merged_df, right=Geography_df, how='left', left_on='LOGRECNO', right_on='Logical Record Number')
merged_df.drop(merged_df.filter(regex='_y$|_x$|Logical Record Number').columns.tolist(),axis=1, inplace=True)
merged_df.columns = merged_df.columns.str.replace('%',' ')
```

Getting the location of FILEID and next few consecutive non-required colummns

```python
merged_df.columns.get_loc('FILEID')
```

Dropping the FILEID and few not required columns from the merged df

```python
merged_df= merged_df.drop(merged_df.columns[43:49], axis=1)
#merged_df= merged_df.drop(merged_df.columns[41], axis=1)
#Race_df= Race_df.drop(Race_df.columns[17:], axis=1)
merged_df.head()
```

Getting the columns of the merged df,saving it as list in cols variable and rearrange the columns using slicing.

```python
cols= merged_df.columns.tolist()
cols= cols[-3:]+ cols[:-3]
cols
```

Inputting the rearranged columns into new dataframe

```python
merged_df1= merged_df[cols]
merged_df1.head(10)
```
